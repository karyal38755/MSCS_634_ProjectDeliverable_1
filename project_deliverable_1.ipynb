```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Reading the dataset and analyzing it
df = pd.read_csv('student-mat.csv', sep=';')
print(df.info())
print(df.head())
```

```python
# Checking missing data
print(df.isnull().sum())
```

```python
# Checking duplicate entry
duplicates = df[df.duplicated()]
print(duplicates)
```

```python
# Inspecting inconsistent label
print(df['Mjob'].unique())
```

```python
# Identifying noisy data
sns.boxplot(x=df['absences'])
```

```python
# Applying percentile capping
q_high = df['absences'].quantile(0.95)
df['absences'] = df['absences'].clip(upper=q_high)
```

```python
# Distribution analysis
sns.histplot(df['G3'], bins=20, kde=True)
plt.title("Distribution of Final Grades (G3)")
```

```python
# Categorial feature analysis
sns.countplot(x='schoolsup', hue='sex', data=df)
```

```python
# Inspecting feature correlation
corr = df[['G1', 'G2', 'G3']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python
# Outlier identification
sns.boxplot(x=df['G3'])
```
